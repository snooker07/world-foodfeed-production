## Define  the Problem

Variables in the column headers. Year variables are separated in different columns, it should be one. Melting will solve the problem. After that, delete the original year columns from the table.

## Code

```python
# Gather the erronous column names
erronous_column_names = list(df_clean)[11:]
correct_column_names = list(df_clean)[0:10]
```

```python
# Melting the variable columns into one
df_supply_clean = pd.melt(df_clean, id_vars = "SupplyId", value_vars = erronous_column_names, var_name = "Year", value_name = "Supply")
```

Every unit is in 1000 tonnes. It is redundant information, so we can write this into the column name

## Define the Problem

Because every unit is in 1,000 tonnes, it would be better to have this information in the Supply column header. Ten delete the Unit column.

## Code

```python
# Rename Supply column header
df_supply_clean.rename(columns =
                      {
                          "Supply" : "Supply1000Tonnes"
                      }, inplace = True)
```

```python
# Delete Unit column
df_clean.drop( columns = "Unit", axis = 1, inplace = True)
```

## Define the Problem

Year should be numerical and there is an Y letter in the value. Delete first the Y letter, then change its type to in64.

## Code

```python
# Delete Y letter
df_supply_clean["Year"] = df_supply_clean["Year"].str[1:]
```

```python
# Convert from string to integer
df_supply_clean["Year"] = df_supply_clean["Year"].astype("int64")
```

```python
# Files to save
files_to_save = [csv_supply_info, csv_supply]
```

```python
#Compress files into a zipfile
with zipfile.ZipFile(zip_name, "w") as _zip:
    for i in range(len(files_to_save)):
        _zip.write(files_to_save[i])
```
